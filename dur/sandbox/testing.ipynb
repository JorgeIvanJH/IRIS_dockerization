```python
import os
import pandas as pd
```

```python


def adjust_datasize(datapath, NUM_ROWS):
    """
    Meant ONLY for testing purposes.
    Adjust the dataframe to have exactly NUM_ROWS rows by repeating or subsampling.
    """
    path, filename = os.path.split(datapath)
    filename, ext = os.path.splitext(filename)
    df = pd.read_csv(datapath)
    n_rows = df.shape[0]
    nrepeats,remiainder = NUM_ROWS // n_rows , NUM_ROWS % n_rows
    df = pd.concat([df]*nrepeats + [df.sample(remiainder)], ignore_index=True)
    new_datapath = os.path.join(path, f"{filename}_nr{NUM_ROWS}{ext}")
    df.to_csv(new_datapath, index=False)
    return new_datapath

```
